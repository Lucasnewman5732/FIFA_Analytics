```python
#We still have a problem:
#We need to convert all our positional data that is str(True-False) to Boolean type, then 
# to dummy integer type where True=1 and False=0

#RW&RM&RWB&RB
complete.prefers_rw = complete.prefers_rw.astype(bool)
complete.prefers_rw = complete.prefers_rw.astype(int)
complete.prefers_rm = complete.prefers_rm.astype(bool)
complete.prefers_rm = complete.prefers_rm.astype(int)
complete.prefers_rb = complete.prefers_rb.astype(bool)
complete.prefers_rb = complete.prefers_rb.astype(int)
complete.prefers_rwb = complete.prefers_rwb.astype(bool)
complete.prefers_rwb = complete.prefers_rwb.astype(int)
#ST
complete.prefers_st = complete.prefers_st.astype(bool)
complete.prefers_st = complete.prefers_st.astype(int)
#LW&LM&LWB&LB
complete.prefers_lw = complete.prefers_lw.astype(bool)
complete.prefers_lw = complete.prefers_lw.astype(int)
complete.prefers_lm = complete.prefers_lm.astype(bool)
complete.prefers_lm = complete.prefers_lm.astype(int)
complete.prefers_lb = complete.prefers_lb.astype(bool)
complete.prefers_lb = complete.prefers_lb.astype(int)
complete.prefers_lwb = complete.prefers_lwb.astype(bool)
complete.prefers_lwb = complete.prefers_lwb.astype(int)

#CM&CDM&CAM&CF
complete.prefers_cm = complete.prefers_cm.astype(bool)
complete.prefers_cm = complete.prefers_cm.astype(int)
complete.prefers_cdm = complete.prefers_cdm.astype(bool)
complete.prefers_cdm = complete.prefers_cdm.astype(int)
complete.prefers_cam = complete.prefers_cam.astype(bool)
complete.prefers_cam = complete.prefers_cam.astype(int)
complete.prefers_cf = complete.prefers_cf.astype(bool)
complete.prefers_cf = complete.prefers_cf.astype(int)

#CB
complete.prefers_cb = complete.prefers_cb.astype(bool)
complete.prefers_cb = complete.prefers_cb.astype(int)

#GK
complete.prefers_gk = complete.prefers_gk.astype(bool)
complete.prefers_gk = complete.prefers_gk.astype(int)
complete.head()
```

```python
#Addressing Missing Values
#Do we have missing values
#If we do fill them with zeros along axis=0
complete.isnull().any(axis=0)
complete.fillna(0, inplace=True, axis=0)
```

```python
#Check if we got rid of missing data
complete.isnull().any(axis=0)
complete.isnull().sum(axis=0)
```

```python
#connect to sql db
conn = sqlite3.connect(database)
cur = conn.cursor()
tables = pd.read_sql("""SELECT * FROM sqlite_master WHERE type='table';""", conn)
```

```python
#Create league df and modify it, then export as csv
leagues = pd.read_sql("""SELECT *
                        FROM League
                        JOIN Country ON Country.id = League.country_id;""", conn,)

leagues=leagues.drop(['id'], axis=1)
leagues.set_index('country_id')
leagues.to_csv('Euro_leagues.csv', encoding="utf-8")
```

```python
#create teams table
teams = pd.read_sql("""SELECT *
                        FROM Team
                        ORDER BY team_long_name;""", conn)

teams=teams.drop(['id'], axis=1)
teams=teams.drop(['team_fifa_api_id'], axis=1)
teams=teams.set_index('team_api_id')
teams.head()
teams.to_csv('Euro_teams.csv', encoding="utf-8")
```

```python
#create a team attributes table
#this will be important when trying to figure out what players fit a scheme of a team better
team_attributes = pd.read_sql("""SELECT 
                            * FROM Team_Attributes AS theleft 
                            LEFT OUTER JOIN Team AS theright 
                            ON theleft.team_api_id=theright.team_api_id 
                            ORDER BY team_long_name;""", conn)
#Lets see how many null values
team_attributes.isnull().sum(axis=0)
#As we can see 'buildUpPlayDribbling' has a lot of nulls and is the only column that has any
#Let us drop this column

team_attributes=team_attributes.drop(['buildUpPlayDribbling'], axis=1)
#lets check again
team_attributes.isnull().sum(axis=0)
team_attributes.head()

team_attributes.to_csv('Euro_team_attributes.csv', encoding="utf-8")
```

```python
detailed_matches = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name = 'Spain'
                                ORDER by date""", conn)
detailed_matches
detailed_matches.to_csv('/Users/lucasnewman/Desktop/FIFA/detailed_matches.csv', encoding='utf-8')
```

```python
#Create a player reference table
player_ref=pd.read_sql("""SELECT *
                        FROM Player
                        ORDER BY player_fifa_api_id ;""", conn)
player_ref.isnull().sum(axis=0)
player_ref=player_ref.drop(['id'], axis=1)
player_ref=player_ref.drop(['player_api_id'], axis=1)
player_ref=player_ref.set_index('player_fifa_api_id')
player_ref.head()
player_ref.to_csv('player_ref.csv', encoding="utf-8")
```

```python
#create a table with only player attributes and id
just_player_attributes=pd.read_sql("""SELECT *
                        FROM Player_Attributes
                        ORDER BY player_fifa_api_id ;""", conn)
just_player_attributes.head()
just_player_attributes.isnull().sum(axis=0)
just_player_attributes=just_player_attributes.dropna()
just_player_attributes.isnull().sum(axis=0)
just_player_attributes=just_player_attributes.drop(['id'], axis=1)
just_player_attributes=just_player_attributes.drop(['player_api_id'], axis=1)
just_player_attributes=just_player_attributes.set_index('player_fifa_api_id')
just_player_attributes.head()
just_player_attributes.to_csv('just_player_attributes.csv', encoding="utf-8")
```

```python
#Join player reference and player attribute tables so that now player attribute table 
#has names
left=just_player_attributes
right=player_ref
euro_player_df=left.join(right, lsuffix='_left', rsuffix='_right')
euro_player_df.head()
euro_player_df.isnull().sum(axis=0)
euro_player_df.to_csv('euro_player.csv', encoding="utf-8")
```
